| В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame(random.sample(['robot', 'human']*10, 10) ,columns={'Colors'})
data.head(10)
```

```python
data.loc[data['Colors'] == 'human', 'captcha'] = '1'
data.loc[data['Colors'] == 'robot', 'captcha'] = '0'
data.head(10)
```
